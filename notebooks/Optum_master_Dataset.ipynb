# Analysis

```python
df_patients["BIRTHDATE"]=df_patients["BIRTHDATE"].apply(lambda x : int(x[0:4]))
df_patients["DEATHDATE"]=df_patients["DEATHDATE"].apply(lambda x : int(x[0:4]) if type(x)==str else 2021)
```

```python
df_patients["Age"]=df_patients["DEATHDATE"]-df_patients["BIRTHDATE"]
```

```python
patients = list(df_patients["Id"])
print(patients)
```

```python
# distribution in patients
patient_distribution = dict(df_medications["PATIENT"].value_counts())
plt.bar(range(len(patient_distribution)),list(patient_distribution.values()),tick_label=list(patient_distribution.keys()))
plt.xticks([])
plt.show()
```

```python
# columns needed from patients csv
master_dict={}
master_dict["ID"]=list(df_patients["Id"])
master_dict["BIRTHYEAR"]=list(df_patients["BIRTHDATE"])
master_dict["AGE"]=list(df_patients["Age"])
master_dict["MARITAL"]=list(df_patients["MARITAL"])
master_dict["RACE"]=list(df_patients["RACE"])
master_dict["ETHNICITY"]=list(df_patients["ETHNICITY"])
master_dict["GENDER"]=list(df_patients["GENDER"])
master_dict["STATE"]=list(df_patients["STATE"])
master_dict["CITY"]=list(df_patients["CITY"])
master_dict["COUNTY"]=list(df_patients["COUNTY"])
master_dict["HEALTHCARE_EXPENSES"]=list(df_patients["HEALTHCARE_EXPENSES"])
master_dict["HEALTHCARE_COVERAGE"]=list(df_patients["HEALTHCARE_COVERAGE"])
medication_count = []
for patient in master_dict["ID"]:
  if patient in patient_distribution:
    medication_count.append(patient_distribution[patient])
  else:
    medication_count.append(0)
master_dict["MEDICATION_COUNT"]=medication_count
```

```python
for key in master_dict.keys():
  print(key,len(master_dict[key]))
```

```python
master_dataset = pd.DataFrame(master_dict)
master_dataset.to_csv("/content/drive/MyDrive/csv/master.csv")
```
